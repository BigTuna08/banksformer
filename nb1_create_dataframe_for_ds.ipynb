# Create dataframe with all unscaled fields

```python
ds_suffix = "vf10k"
input_df = "data/custom_txns_10k.csv"
```

```python
df = pd.read_csv(input_df, parse_dates=["dateofbirth","transactiondate"])
df
```

```python
df["transactionactivitytype"].fillna("", inplace=True)
df
```

```python
df["age"] = 2021- df["dateofbirth"].dt.isocalendar()["year"]
df
```

```python
df["direction"] = df["totalvalue"].apply(lambda x: "db" if x < 0 else "cr")
df
```

```python
" + '-' + ".join(["df." + x for x in 
            ["transactiontype", "transactionmethodtype", "sourcefundtype", "destinationfundtype", "transactionactivitytype"]])
```

```python
df["tcode"] = df.transactiontype + '-' + df.transactionmethodtype + '-' +\
                            df.sourcefundtype + '-' + df.destinationfundtype + '-' +\
                            df.transactionactivitytype + "-" + df["direction"]

df
```

```python
df["amount"] = np.abs(df["totalvalue"])
df
```

```python
df.rename(columns={"customerid": "account_id"}, inplace=True)
```

```python
df["datetime"] = df["transactiondate"].apply(lambda x: x.date()) # ignore time info for consistency with czech data (can add in later)
```

```python
df
```

```python
df["month"] = df["transactiondate"].dt.month
df["day"] = df["transactiondate"].dt.day
df["dow"] =  df["transactiondate"].dt.dayofweek
df["year"] = df["transactiondate"].dt.year
df
```

```python
df["td"] = df[["account_id", "datetime"]].groupby("account_id").diff()

def days_from_timedelta(td):
    try: return td.days
    except: return 0.0

df["td"] = df["td"].apply(days_from_timedelta)
df
```

```python
df.to_csv(f"stored_data/final_df-{ds_suffix}.csv", index=False)
```
